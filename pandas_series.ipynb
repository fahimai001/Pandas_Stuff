```python
# Using sigmoid() find the sigmoid value of the following
temp = pd.DataFrame([10,20,30,40,50])
```

```python
def sigmoid(fahim):
    return 1/1+np.exp(-fahim)
```

```python
temp['Sigmoid_value'] = temp.apply(sigmoid)
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    
    return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['Euclidean_Distance'] = points_df.apply(euclidean, axis=1)
```
